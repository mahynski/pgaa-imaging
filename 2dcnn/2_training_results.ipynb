```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
import watermark
%load_ext watermark

%load_ext autoreload
%autoreload 2

%watermark -t -m -v --iversions
```

```python
h1 = pickle.load(open('model1_history.pkl', 'rb'))
h2 = pickle.load(open('model2_history.pkl', 'rb'))
h3 = pickle.load(open('model3_history.pkl', 'rb'))
```

```python
steps = np.arange(1, len(h1['loss'])+1)

fig, ax = plt.subplots(nrows=2, ncols=1, figsize=(4,8))

ax[0].plot(
    steps,
    h1['val_sparse_categorical_accuracy'],
    label='Model 1 (MobileNetV3Small)'
)

ax[0].plot(
    steps,
    h2['val_sparse_categorical_accuracy'],
    label='Model 2 (NasNetMobile)'
)

ax[0].plot(
    steps,
    h3['val_sparse_categorical_accuracy'],
    label='Model 3 (InceptionV3)'
)

ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Validation Accuracy')
ax[0].legend(loc='best')

ax[1].plot(
    steps,
    h1['val_loss'],
    label='Model 1 (MobileNetV3Small)'
)

ax[1].plot(
    steps,
    h2['val_loss'],
    label='Model 2 (NasNetMobile)'
)

ax[1].plot(
    steps,
    h3['val_loss'],
    label='Model 3 (InceptionV3)'
)

ax[1].set_yscale('log')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Validation Loss')

# _ = plt.savefig('training_performance.png', dpi=300, bbox_inches='tight')
```
