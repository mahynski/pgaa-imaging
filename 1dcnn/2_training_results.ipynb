# Utiles

```python
def plot_histories(train_time_history, loc1=(0.25, 0.8), loc2=(1, 0.5), ylim=None, filename=None, figsize=(5, 5), fontsize=10):
    fig, ax1 = plt.subplots(figsize=figsize)
    
    val_x = np.arange(1, len(train_time_history['accuracy'])+1)
    train_x = val_x - 0.5
    ax1.plot(train_x, train_time_history['sparse_categorical_accuracy'], '-', color='C0', label='Training Accuracy', alpha=0.5)
    ax1.plot(val_x, train_time_history['val_sparse_categorical_accuracy'], '-', color='C1', label='Validation Accuracy', alpha=0.5)
    
    ax1.set_ylim(ylim)
    ax1.set_xlabel('Epoch', fontsize=fontsize)
    ax1.set_ylabel('Categorical Accuracy', fontsize=fontsize)
    ax1.legend(bbox_to_anchor=loc1, fontsize=fontsize-4)
    
    ax1.set_xticklabels(ax1.get_xticklabels(), fontsize=fontsize-4)
    ax1.set_yticklabels(ax1.get_yticklabels(), fontsize=fontsize-4)
    
    ax2 = ax1.twinx()
    ax2.plot(train_x, train_time_history['loss'], '-', color='C2', label='Training Loss', alpha=0.5)
    ax2.plot(val_x, train_time_history['val_loss'], '-', color='C3', label='Validation Loss', alpha=0.5)
    
    ax2.set_ylabel('Loss', fontsize=fontsize)
    ax2.legend(bbox_to_anchor=loc2, fontsize=fontsize-4)
    
    ax2.set_xticklabels(ax2.get_xticklabels(), fontsize=fontsize-4)
    ax2.set_yticklabels(ax2.get_yticklabels(), fontsize=fontsize-4)
    
    if filename is not None:
        plt.tight_layout()
        plt.savefig(fname=filename, transparent=None, dpi=300, format=None,
            metadata=None, bbox_inches=None, pad_inches=0.1,
            facecolor='auto', edgecolor='auto', backend=None
        )
        
    return ax1
```
